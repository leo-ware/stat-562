**2. Suppose that $X$ has the geometric distribution with parameter $p$. (See Sec. 5.5.) Find the Fisher information $I(p)$ in $X$.**

In the case where $p \in \{0, 1\}$, the support of $X$ depends on $p$. So, we will assume $p \in (0, 1)$.

Assume $X \in \{0, 1, 2, ...\}$. If $X$ has the geometric distribution, then,

$$
P(X = x) = p(1 - p)^x
$$

Then, the log likelihood is given by,

$$
L(p) = \ln(p(1 - p)^X)
$$

$$
L(p) = \ln(p) + X\ln(1 - p)
$$

Then,

$$
{\partial \over \partial p}L(p) = {\partial \over \partial p}(\ln(p) + x\ln(1 - p))
$$

$$
{\partial \over \partial p}L(p) = p^{-1} - X(1 - p)^{-1}
$$

Note that this function is differentiable with respect to $p$.

Then,

$$
var({\partial \over \partial p}L(p)) = var(p^{-1} - X(1 - p)^{-1})
$$

But, $p^{-1}$ is constant with respect to $X$. So,

$$
var({\partial \over \partial p}L(p)) = var(-(1 - p)^{-1}X)
$$

$$
var({\partial \over \partial p}L(p)) = (-(1 - p)^{-1})^2var(X)
$$

$$
var({\partial \over \partial p}L(p)) = (1 - p)^{-2}var(X)
$$

The variance of the geometric distribution is known.

$$
var({\partial \over \partial p}L(p)) = (1 - p)^{-2}((1 - p)p^{-2})
$$

$$
var({\partial \over \partial p}L(p)) = (1 - p)^{-1}p^{-2}
$$

We have that the support of $X$ does not depend on $p$, $p$ lies on an open interval, and the log likelihood is twice differentiable.So,

$$
I(p) = {1 \over (1 - p)p^2}
$$

**3. Suppose that a random variable $X$ has the Poisson distribution with unknown mean $θ > 0$. Find the Fisher information $I(θ)$ in $X$.**

Assume $X \in \{0, 1, 2, ...\}$. If $X$ has the Poisson distribution, then,

$$
P(X = x) = {e^{-\theta}\theta^x \over x!}
$$

Note that the support of $X$ is unaffected by $\theta$. Note that $\theta$ lies on the open interval $(0, \infty)$.

Then, the log likelihood is given by,

$$
L(\theta) = \ln\left( {e^{-\theta}\theta^x \over x!} \right)
$$

$$
L(\theta) = \ln(e^{-\theta}) + \ln(\theta^x) + \ln(x!^{-1})
$$

$$
L(\theta) = -\theta + x\ln(\theta) - \ln(x!)
$$

Then,

$$
{d \over d\theta}L(\theta) ={d \over d\theta}( -\theta + x\ln(\theta) - \ln(x!))
$$

$$
{d \over d\theta}L(\theta) = x\theta^{-1} - 1
$$

Note that this function is differentiable with respect to $\theta$.

Then,

$$
var({d \over d\theta}L(\theta)) = var(x\theta^{-1} - 1)
$$

Variance is translation invariant.

$$
var({d \over d\theta}L(\theta)) = var(x\theta^{-1})
$$

The parameter $\theta$ is constant with respect to $x$.

$$
var({d \over d\theta}L(\theta)) = \theta^{-2}var(x)
$$

The variance of the poisson distribution is known.

$$
var({d \over d\theta}L(\theta)) = \theta^{-2}\theta
$$

$$
var({d \over d\theta}L(\theta)) = \theta^{-1}
$$

We have that the support of $X$ does not depend on $\theta$, $\theta$ lies on an open interval, and the second derivative of the log likelihood exists. So,

$$
I(\theta) = {1 \over \theta}
$$

**4. Suppose that a random variable has the normal distribution with mean $0$ and unknown standard deviation $σ > 0$. Find the Fisher information $I(σ)$ in $X$.**

Assume $X \in \mathbb{R}$. If $X$ has the normal distribution with $\mu = 0$, then,

$$
P(X = x) = {1 \over \sqrt{2\pi \sigma^2}} \exp\!\left(-{x^2 \over 2\sigma^2}\right)
$$

Note that the support of $X$ does not depend on $\sigma$. Note that $\sigma$ lies on the open interval $(0, \infty)$.

Then, the log likelihood of $\sigma$ given $X$ is given by,

$$
L(\sigma) = \ln\left({1 \over \sqrt{2\pi \sigma^2}} \exp\!\left(-{X^2 \over 2\sigma^2}\right) \right)
$$

$$
L(\sigma) = \ln\left((\sqrt{2\pi} \sigma)^{-1}\right) + \ln\left( \exp\!\left(-{X^2 \over 2\sigma^2}\right) \right)
$$

$$
L(\sigma) =
-\ln(\sqrt{2\pi})
-\ln(\sigma)
-{X^2 \over 2\sigma^2}
$$

Then,

$$
{d \over d\sigma} L(\sigma) =
{d \over d\sigma}(
-\ln(\sqrt{2\pi})
-\ln(\sigma)
-{X^2\sigma^{-2} \over 2}
)$$

$$
{d \over d\sigma} L(\sigma) =
X^2\sigma^{-3}
-\sigma^{-1}
$$

Note that this function is differentiable with respect to $\sigma$.

Then,

$$
var({d \over d\sigma} L(\sigma)) = var(X^2\sigma^{-3} - \sigma^{-1})
$$

Variance is translation invariant. The parameter $\sigma$ is constant with respect to $X$.

$$
var({d \over d\sigma} L(\sigma)) = var(X^2\sigma^{-3})
$$

$$
var({d \over d\sigma} L(\sigma)) = \sigma^{-2}var(X^2\sigma^{-2})
$$

$$
var({d \over d\sigma} L(\sigma)) = \sigma^{-2}var((X/\sigma)^{2})
$$

The random variable $x/\sigma$ is a linear transform of a normal variable. So, it is a normal variable.

$$
E[X/\sigma] = {1 \over \sigma}E[X] = 0
$$

$$
var[X/\sigma] = {1 \over \sigma^2}var[X] = {\sigma^2 \over \sigma^2} = 1
$$

So,

$$
X/\sigma \sim \text{Normal}(0, 1)
$$

So,

$$
(X/\sigma)^2 \sim \chi^2(1)
$$

The variance of the $\chi^2$ is known.

$$
var((X/\sigma)^2) = 2
$$

So,

$$
var({d \over d\sigma} L(\sigma)) = \sigma^{-2}2
$$

We have that the support of $X$ does not depend on $\sigma$, $\sigma$ lies on an open interval, and the second derivative of the log likelihood exists. So,

$$
I(\sigma) = {2 \over \sigma^2}
$$

**5. Suppose that a random variable $X$ has the normal distribution with mean 0 and unknown variance $σ^2 > 0$. Find the Fisher information $I(σ^2)$ in $X$. Note that in this exercise the variance $σ^2$ is regarded as the parameter, whereas in Exercise 4 the standard deviation $σ$ is regarded as the parameter.**

Assume $X \in \mathbb{R}$. If $X$ has the normal distribution with $\mu = 0$, then,

$$
P(X = x) = {1 \over \sqrt{2\pi \sigma^2}} \exp \left(-{x^2 \over 2\sigma^2}\right)
$$

Note that the support of $X$ does not depend on $\sigma^2$. Note that $\sigma^2$ lies on the open interval $(0, \infty)$.

Let $\phi = \sigma^2$,

$$
P(X = x) = (2\pi\phi)^{-1/2} \exp \left(-{x^2 \over 2}\phi^{-1}\right)

The the log likelihood is given by,

$$
L(\phi) = \ln((2\pi\phi)^{-1/2} \exp \left(-{x^2 \over 2}\phi^{-1}\right))
$$

$$
L(\phi) = -(1/2)\ln(2\pi\phi) -{x^2 \over 2}\phi^{-1}
$$

$$
L(\phi) = -(1/2)(\ln(2\pi) + \ln(\phi) + x^2\phi^{-1})

Then,

$$
{d\over d\phi}L(\phi) = -(1/2){d\over d\phi}(\ln(2\pi) + \ln(\phi) + x^2\phi^{-1})
$$

$$
{d\over d\phi}L(\phi) = -(1/2)(\phi^{-1} - x^2\phi^{-2})
$$

$$
{d\over d\phi}L(\phi) = (1/2)(x^2\phi^{-2} - \phi^{-1})
$$

Note this function is differentiable with respect to $\phi$.

Then,

$$
var({d\over d\phi}L(\phi)) = var((1/2)(x^2\phi^{-2} - \phi^{-1}))
$$

By the properties of variance,

$$
var({d\over d\phi}L(\phi)) = (1/2)^2var(x^2\phi^{-2})
$$

$$
var({d\over d\phi}L(\phi)) = (1/2)^2var(x^2\phi^{-1}\phi^{-1})

var({d\over d\phi}L(\phi)) = (1/2)^2\phi^{-2}var(x^2\phi^{-1})

We found above that,

var({d\over d\phi}L(\phi)) = {2 \over 4\phi^2}

We have that the support of $X$ does not depend on $\phi$, $\phi$ lies on an open interval, and the second derivative of the log likelihood exists. So,

I(\phi) = {1 \over 2\sigma^4}

**7. Suppose that $X_1, . . . , X_n$ form a random sample from the Bernoulli distribution with unknown parameter $p$. Show that $\overline{X}_n$ is an efficient estimator of $p$.**

*Case 1*: $p \in \{0, 1\}$

If $p \in \{0, 1\}$, then $\overline{X}_n = p$ and $var(\overline{X}_n) = 0$. So, in this case, the sample mean is unbiased and minimum variance.

*Case 2*: $p \in (0, 1)$

Note that, in this case, $p$ lies on an open interval. Note that, in this case, $p \ne 0$ and $(1 - p) \ne 0$. Note that, in this case, the support of $X$ does not depend on $p$.

We have that,

$$
P(X_i = x | p) = p^x(1 - p)^{1 - x}
$$

So, the log likelihood of $p$ is given by,

$$
L(p) = \ln(p^x(1 - p)^{1 - x})
$$

$$
L(p) = x\ln(p) + (1 - x)\ln(1 - p)
$$

$$
{d \over dp}L(p) = {d \over dp}(x\ln(p) + (1 - x)\ln(1 - p))
$$

$$
{d \over dp}L(p) = xp^{-1} - (1 - x)(1 - p)^{-1}
$$

$$
{d \over dp}L(p) = xp^{-1} + x(1 - p)^{-1} - (1 - p)^{-1}
$$

$$
{d \over dp}L(p) = x(p^{-1} + (1 - p)^{-1}) - (1 - p)^{-1}
$$

Note that this function is differentiable with respect to $p$.

$$
var({d \over dp}L(p)) = var(x(p^{-1} + (1 - p)^{-1}) - (1 - p)^{-1})
$$

By the properties of variance,

$$
var({d \over dp}L(p)) = (p^{-1} + (1 - p)^{-1})^2var(x)
$$

The variance of the Bernoulli is known.

$$
var({d \over dp}L(p)) = (p^{-1} + (1 - p)^{-1})^2(p(1 - p))
$$

$$
var({d \over dp}L(p)) = (p^{-2} + 2p^{-1}(1 - p)^{-1} + (1 - p)^{-2})(p(1 - p))
$$

$$
var({d \over dp}L(p)) = (p(1 - p))p^{-2} + (p(1 - p))2p^{-1}(1 - p)^{-1} + (p(1 - p))(1 - p)^{-2}
$$

$$
var({d \over dp}L(p)) = (1 - p)p^{-1} + 2 + p(1 - p)^{-1}
$$

$$
var({d \over dp}L(p)) = {(1 - p) \over p} + 2 + {p \over (1 - p)}
$$

$$
var({d \over dp}L(p)) = {(1 - p)^2 \over p(1 - p)} + {2p(1 - p) \over p(1 - p)} + {p^2 \over p(1 - p)}
$$

$$
var({d \over dp}L(p)) = {(1 - p)^2 + p^2 + 2p(1 - p) \over p(1 - p)}
$$

$$
var({d \over dp}L(p)) = {1 - 2p + p^2 + p^2 + 2p - 2p^2 \over p(1 - p)}
$$

$$
var({d \over dp}L(p)) = {1 \over p(1 - p)}
$$

We have that, in this case, the support of $X$ does not depend on $p$, $p$ is on an open interval, and the second derivative of the log likelihood exists. Then, for the case of a single Bernoulli variable,

$$
I(p) = {1 \over p(1 -p)}
$$

So, for the case of a random sample of $n$ Bernoulli variables,

$$
I(p) = {n \over p(1 -p)}
$$

So, by the Cramer-Rao bound, the lowest possible variance of an unbiased estimator is,

$$
I^{-1}(p) = {1 \over n}p(1 -p)
$$

Consider the sample mean,

$$
\overline{X}_n = {1 \over n}\sum_{i = 1}^nX_i
$$

Take the expectation.

$$
E[\overline{X}_n] = E[{1 \over n}\sum_{i = 1}^nX_i]
$$

$$
E[\overline{X}_n] = {1 \over n}\sum_{i = 1}^nE[X_i]
$$

The expectation of the Bernoulli is known.

$$
E[\overline{X}_n] = {1 \over n}\sum_{i = 1}^np
$$

$$
E[\overline{X}_n] = p
$$

Take the variance.

$$
var(\overline{X}_n) = var({1 \over n}\sum_{i = 1}^nX_i)
$$

$$
var(\overline{X}_n) = {1 \over n^2}var(\sum_{i = 1}^nX_i)
$$

The $X_i$ are independent.

$$
var(\overline{X}_n) = {1 \over n^2}\sum_{i = 1}^nvar(X_i)
$$

The variance of the Bernoulli is known.

$$
var(\overline{X}_n) = {1 \over n^2}\sum_{i = 1}^np(1 - p)
$$

$$
var(\overline{X}_n) = {1 \over n^2}np(1 - p)
$$

$$
var(\overline{X}_n) = {1 \over n}p(1 - p)
$$

So, in this case, the variance of the sample mean is equal to the minimum variance given by the Cramer-Rao bound. Further, it is unbiased. We can conclude that the sample mean is an efficient estimator of $p$.

So, in either case, the estimator is efficient and unbiased.

**8. Suppose that $X_1, . . . , X_n$ form a random sample from the normal distribution with unknown mean $μ$ and known variance $σ^2 > 0$. Show that $X_n$ is an efficient estimator of $μ$.**

We have that,

$$
P(X_i = x) = {1 \over \sqrt{2\pi \sigma^2}} \exp \left(-{(x - \mu)^2 \over 2\sigma^2}\right)
$$

Note that the support of $X_i$ is unaffected by $\mu$. Note that $\mu$ can take on any value in the open interval $(-\infty, \infty)$.

So, the log likelihood of $\mu$ is given by,

$$
L(\mu) = \ln({1 \over \sqrt{2\pi \sigma^2}} \exp \left(-{(x - \mu)^2 \over 2\sigma^2}\right))
$$

$$
L(\mu) = \ln({1 \over \sqrt{2\pi \sigma^2}}) + \left(-{(x - \mu)^2 \over 2\sigma^2}\right)
$$

Then,

$$
{d \over d\mu}L(\mu) = {d \over d\mu}(\ln({1 \over \sqrt{2\pi \sigma^2}}) + \left(-{(x - \mu)^2 \over 2\sigma^2}\right))
$$

$$
{d \over d\mu}L(\mu) = {d \over d\mu}-{(x - \mu)^2 \over 2\sigma^2}
$$

$$
{d \over d\mu}L(\mu) = -{2(x - \mu) \over 2\sigma^2}(-1)
$$

$$
{d \over d\mu}L(\mu) = {x - \mu \over \sigma^2}
$$

Note this function is differentiable with respect to $\mu$.

Then,

$$
var({d \over d\mu}L(\mu)) = var({x - \mu \over \sigma^2})
$$

By the properties of variance.

$$
var({d \over d\mu}L(\mu)) = {1 \over \sigma^4}var(x - \mu)
$$

By the properties of variance.

$$
var({d \over d\mu}L(\mu)) = {1 \over \sigma^4}var(x)
$$

The variance of the normal is known.

$$
var({d \over d\mu}L(\mu)) = {1 \over \sigma^4}\sigma^2
$$

$$
var({d \over d\mu}L(\mu)) = \sigma^{-2}
$$

We have that the support of $X_i$ does not depend on $\mu$, $\mu$ lies on an open interval, and the log likelihood is twice differentiable. So, for a single variable $X_i$.

$$
I(\mu) = \sigma^{-2}
$$

Then, for a random sample $X_1...X_n$.

$$
I(\mu) = n\sigma^{-2}
$$

Then, by the Cramer-Rao bound, the minimum variance of an unbiased estimator is given by,

$$
I^{-1}(\mu) = {\sigma^{2} \over n}
$$

Consider the sample variance.

$$
\overline{X}_n = {1 \over n}\sum_{i = 1}^n X_i
$$

Take the expectation.

$$
E[\overline{X}_n] = E[{1 \over n}\sum_{i = 1}^n X_i]
$$

The $X_i$ are independent, so by the properties of expectation,

$$
E[\overline{X}_n] = {1 \over n}\sum_{i = 1}^n E[X_i]
$$

The expectation of a normal is known.

$$
E[\overline{X}_n] = {1 \over n}\sum_{i = 1}^n \mu
$$

Then,

$$
E[\overline{X}_n] = {1 \over n}n \mu
$$

$$
E[\overline{X}_n] = \mu
$$

So, the sample mean is an unbiased estimator for $\mu$.

Take the variance.

$$
var[\overline{X}_n] = var[{1 \over n}\sum_{i = 1}^n X_i]
$$

By the properties of variance,

$$
var[\overline{X}_n] = {1 \over n^2}var[\sum_{i = 1}^n X_i]
$$

The $X_i$ are independent, so by the properties of variance,

$$
var[\overline{X}_n] = {1 \over n^2}\sum_{i = 1}^n var[X_i]
$$

The variance of the normal is known.

$$
var[\overline{X}_n] = {1 \over n^2}\sum_{i = 1}^n \sigma^2
$$

$$
var[\overline{X}_n] = {1 \over n^2}n \sigma^2
$$

$$
var[\overline{X}_n] = {1 \over n} \sigma^2
$$

So, the sample mean is unbiased, and the variance is equal to the minimum variance given by the Cramer-Rao bound for unbiased estimators. So, the sample mean is efficient.

**10. Suppose that $X_1, . . . , X_n$ form a random sample from the normal distribution with mean $0$ and unknown standard deviation $σ > 0$. Find the lower bound specified by the information inequality for the variance of any unbiased estimator of $\log σ$.**
