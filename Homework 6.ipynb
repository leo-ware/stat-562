$$

The MLE estimator for the variance, $\hat{\sigma}^2$, is given by,

$$
\hat{\sigma}^2 = {1 \over n}\sum_{i = 1}^n (X_i - \hat{\mu})^2
$$

Using invariance, we can get the MLE for the standard deviation by taking the square root of both sides.

$$
\hat{\sigma} = \sqrt{{1 \over n}\sum_{i = 1}^n (X_i - \hat{\mu})^2}
$$

$$
\hat{\sigma} = {1 \over \sqrt{n}}\sqrt{\sum_{i = 1}^n (X_i - \hat{\mu})^2}
$$

The sample standard deviation is given by,

$$
\sigma' = \sqrt{{1 \over n - 1}\sum_{i = 1}^n (X_i - \hat{\mu})^2}
$$

$$
\sigma' = {1 \over \sqrt{n - 1}}\sqrt{\sum_{i = 1}^n (X_i - \hat{\mu})^2}
$$

So,

$$
\sigma' = {\sqrt{n} \over \sqrt{n - 1}}\hat{\sigma}
$$

We are interested in the event $\hat{\mu} \gt \mu k\hat{\sigma}$.

$$
\hat{\mu} \gt \mu k\hat{\sigma} = \hat{\mu} - \mu \gt k\hat{\sigma}
$$

$$
\hat{\mu} \gt \mu k\hat{\sigma} = {\sqrt{n} \over \sqrt{n - 1}}(\hat{\mu} - \mu) \gt k{\sqrt{n} \over \sqrt{n - 1}}\hat{\sigma}
$$

$$
\hat{\mu} \gt \mu k\hat{\sigma} = {\sqrt{n} \over \sqrt{n - 1}}(\hat{\mu} - \mu) \gt k\sigma'
$$

$$
\hat{\mu} \gt \mu k\hat{\sigma} = {\sqrt{n} \over \sigma'}(\hat{\mu} - \mu) \gt k\sqrt{n - 1}
$$

$$
P(\hat{\mu} \gt \mu k\hat{\sigma}) = P({\sqrt{n} \over \sigma'}(\hat{\mu} - \mu) \gt k\sqrt{n - 1})
$$

We have that,

$$
{\sqrt{n} \over \sigma'}(\hat{\mu} - \mu) \sim T(n - 1)
$$

Let $\Phi$ represent the CDF of the t-distribution with $n - 1$ degrees of freedom. Let $\Phi^{-1}$ represent the corresponding inverse CDF. Then,

$$
P({\sqrt{n} \over \sigma'}(\hat{\mu} - \mu) \gt k\sqrt{n - 1}) = 1 - \Phi(k\sqrt{n - 1})

Then,

$$

$$

$$
L(\theta) = \sum_{i = 1}^n \ln \left( {\theta^{x_i}e^{-\theta} \over x_i!} \right)

$$
L(\theta) = \sum_{i = 1}^n \left( x_i \ln(\theta) - \theta - \ln(x_i!) \right)

$$
L(\theta) =  \left(\sum_{i = 1}^n x_i \ln(\theta)\right) - \left(\sum_{i = 1}^n\theta\right) - \left(\sum_{i = 1}^n \ln(x_i!) \right)

$$
L(\theta) =
\ln(\theta)S_n
- n\theta
- \left(\sum_{i = 1}^n \ln(x_i!) \right)
$$

Take the derivative.

$$
{d \over d\theta} L(\theta) = {S_n \over \theta} - n

$$

$$
{S_n \over \hat{\theta}} = n

\hat{\theta} = {S_n \over n}

\hat{\theta} = {1 \over n}\sum_{i = 1}^n x_i

**5. Suppose that a random variable $X$ has the Poisson distribution with unknown mean $λ$ $(λ > 0)$. Find a statistic $δ(X)$ that will be an unbiased estimator of $e^λ$.**

Suppose $\delta$ is an unbiased estimator for $e^\lambda$. Then,

$$
E[\delta(X)] = e^\lambda
$$

$$
\sum_{x = 0}^\infty \delta(X) {e^{-\lambda} \lambda^x \over x!} = e^\lambda
$$

$$
e^\lambda\sum_{x = 0}^\infty \delta(X) {e^{-\lambda} \lambda^x \over x!} = e^{2\lambda}
$$

$$
\sum_{x = 0}^\infty \delta(X) {e^{-\lambda}e^\lambda \lambda^x \over x!} = e^{2\lambda}
$$

$$
\sum_{x = 0}^\infty {\delta(X) \over x!}\lambda^x = e^{2\lambda}
$$

The function $e^{2\lambda}$ is analytic for $\lambda \in \mathbb{R}$. So, the left hand side must be analytic for $\lambda \in \mathbb{R}$.

Expand the right hand side using the power series.

$$
\sum_{x = 0}^\infty {\delta(X) \over x!}\lambda^x = \sum_{x = 0}^\infty {(2\lambda)^x \over x!}
$$

Then,

$$
\sum_{x = 0}^\infty {\delta(X) \over x!}\lambda^x = \sum_{x = 0}^\infty {2^x \over x!}\lambda^x
$$

The right hand side is the power series of an analytic function, so it is analytic. The equality hold for $\lambda \in \mathbb{R}$, so by the identity theorem for analytic functions, the coefficients must be equal.

$$
{\delta(X) \over x!} = {2^x \over x!}
$$

$$
\delta(X) = 2^x
$$

So, if $\delta(X)$ is an unbiased estimator for $e^\lambda$, then $\delta(X) = 2^x$. So, if an unbiased estimator exists, then it is $\delta$. We now need to confirm that $2^x$ is indeed an unbiased estimator for $e^\lambda$.

$$
E[2^x] = \sum_{i = 1}^n 2^x {e^{-\lambda}\lambda^x \over x!}
$$

$$
E[2^x] = e^\lambda e^{-\lambda}\sum_{i = 1}^n  {e^{-\lambda}2^x\lambda^x \over x!}
$$

$$
E[2^x] = e^\lambda \sum_{i = 1}^n {e^{-2\lambda}(2\lambda)^x \over x!}
$$

Let $\gamma = 2\lambda$.

$$
E[2^x] = e^\lambda \sum_{i = 1}^n {e^{-\gamma}(\gamma)^x \over x!}
$$

The function,

$$
f_\gamma = {e^{-\gamma}(\gamma)^x \over x!}
$$

is the pdf of the poisson distribution with parameter $\gamma$. So, the sum across the support is 1.

$$
E[2^x] = e^\lambda
$$

So, $\delta(X) = 2^x$ is an unbiased estimator for $e^\lambda$.

**8. Suppose that a random variable $X$ has the geometric distribution with unknown parameter $p$ $(0 < p < 1)$. Show that the only unbiased estimator of $p$ is the estimator $δ(X)$ such that $δ(0)= 1$ and $δ(X)= 0$ for $X > 0$.**

The pdf of the geomtric distribution is given by,

$$
P(X = x | p) = (1 - p)^xp I(x \in \{0, 1, 2,...\})
$$

Suppose $\delta(X)$ is an unbiased estimator of $p$. Then, for all $p$,

$$
E[\delta(X)] = p
$$

$$
\sum_{x = 0}^\infty \delta(x)(1 - p)^xp = p
$$

$$
\sum_{x = 0}^\infty \delta(x)(1 - p)^x = 1
$$

$$
\delta(0) + \sum_{x = 1}^\infty \delta(x)(1 - p)^x = 1
$$

Consider the case where $p = 1$. Then,

$$
\delta(0) + \sum_{x = 1}^\infty \delta(x)(1 - p)^x = \delta(0)
$$

So,

$$
\delta(0) = 1
$$

Note that $\delta$ does not depend on $p$. We have stipulated that $\delta$ is unbiased, so it must be unbiased for any value of $p$, including $p = 1$. So, $\delta(0) = 1$ in general.

Then,

$$
1 + \sum_{x = 1}^\infty \delta(x)(1 - p)^x = 1
$$

$$
\sum_{x = 1}^\infty \delta(x)(1 - p)^x = 0
$$

Let $z = 1 - p$. Then, $z \in [0, 1]$.

$$
\sum_{x = 1}^\infty \delta(x)z^x = 0
$$

The zero function is analytic, so the series must be analytic on $z \in (0, 1)$. By the identity theorem for analytic functions, if two functions are analytic and equal on an open set, they must have the same power series representation on that set. So, they must have the same power series coefficients on that set.

The power series of the zero function is given by,

$$
0 = \sum_{x = 1}^\infty (0)z^x
$$

All the coeffients are zero. So, all the coefficients $\delta(x)$ for $x \ge 1$ must be zero.

So, if $\delta$ is an unbiased estimator for $p$, then,

$$
\delta(x) = \begin{cases}
    1 & \text{for } x = 0 \\
    0 & \text{otherwise}
\end{cases}
$$

We have shown that every unbiased estimator for $p$ is equal to $\delta$. We have not shown there is an unbiased estimator for $p$. We now proceed to show that $\delta$ is an unbiased estimator for $p$.

$$
E[\delta(X)] = \sum_{i = 0}^n \delta(x)(1 - p)^xp
$$

$$
E[\delta(X)] = \delta(0)(1 - p)^(0)p + \sum_{i = 1}^n \delta(x)(1 - p)^xp
$$

Substitute for $\delta$.

$$
E[\delta(X)] = (1)(1 - p)^(0)p + \sum_{i = 1}^n (0)(1 - p)^xp
$$

$$
E[\delta(X)] = p
$$

So, $\delta$ is the unique unbiased estimator for $p$.

**11. Suppose that a certain drug is to be administered to two different types of animals $A$ and $B$. It is known that the mean response of animals of type $A$ is the same as the mean response of animals of type $B$, but the common
value $θ$ of this mean is unknown and must be estimated. It is also known that the variance of the response of animals
of type $A$ is four times as large as the variance of the response of animals of type $B$. Let $X_1, . . . , X_m$ denote the responses of a random sample of m animals of type $A$, and let $Y_1, . . . , Y_n$ denote the responses of an independent random sample of $n$ animals of type $B$. Finally, consider the estimator $\hat{\theta}= α\overline{X}_m + (1− α)\overline{Y}_n$.**

**a. For what values of $α$, $m$, and $n$ is $\hat{\theta}$ an unbiased estimator of $θ$?**

Let,

$$
\phi = var(Y_i)
$$

We have that,

$$
var(X_i) = 4 \phi
$$

$$
E[X_i] = E[Y_i] = \theta
$$

$$
n \ge 0
$$

$$
m \ge 0
$$

Take the expectation of the estimator.

$$
E[\hat{\theta}] = E[\alpha\overline{X}_m + (1 - \alpha)\overline{Y}_n]
$$

$$
E[\hat{\theta}] = \alpha E[\overline{X}_m] + (1 - \alpha)E[\overline{Y}_n]
$$

Case 1: Suppose $n \gt 0$, $m \gt 0$.

$$
E[\hat{\theta}] = \alpha {1 \over m}\sum_{i = 1}^m E[X_i] + (1 - \alpha){1 \over n}\sum_{i = 1}^n E[Y_i]
$$

$$
E[\hat{\theta}] = \alpha {1 \over m}\sum_{i = 1}^m \theta + (1 - \alpha){1 \over n}\sum_{i = 1}^n \theta
$$

$$
E[\hat{\theta}] = \alpha \theta + (1 - \alpha)\theta
$$

$$
E[\hat{\theta}] = (\alpha + (1 - \alpha)) \theta
$$

$$
E[\hat{\theta}] = \theta
$$

So, if $n \gt 0$, $m \gt 0$, $\hat{\theta}$ is unbiased for any $\alpha$.

Case 2: Suppose $n \gt 0$, $m = 0$.

$$
E[\hat{\theta}] = E[\alpha\overline{X}_m + (1 - \alpha)\overline{Y}_n]
$$

$$
E[\hat{\theta}] = E[(1 - \alpha)\overline{Y}_n]
$$

$$
E[\hat{\theta}] = (1 - \alpha)E[{1 \over n}\sum_{i = 1}^n Y_i]
$$

$$
E[\hat{\theta}] = (1 - \alpha){1 \over n}\sum_{i = 1}^n E[Y_i]
$$

$$
E[\hat{\theta}] = (1 - \alpha){1 \over n}\sum_{i = 1}^n \theta
$$

$$
E[\hat{\theta}] = (1 - \alpha) \theta
$$

So, $\hat{\theta}$ is unbiased for $\alpha = 0$.

Case 3: Suppose $m \gt 0$, $n = 0$.

$$
E[\hat{\theta}] = E[\alpha\overline{X}_m + (1 - \alpha)\overline{Y}_n]
$$

$$
E[\hat{\theta}] = E[\alpha\overline{X}_m]
$$

$$
E[\hat{\theta}] = \alpha E[{1 \over m}\sum_{i = 1}^m X_i]
$$

$$
E[\hat{\theta}] = \alpha {1 \over m}\sum_{i = 1}^m E[X_i]
$$

$$
E[\hat{\theta}] = \alpha {1 \over m}\sum_{i = 1}^m \theta
$$

$$
E[\hat{\theta}] = \alpha \theta
$$

So, $\hat{\theta}$ is unbiased for $\alpha = 1$.

So, the choice of $\alpha$ depends on $m$ and $n$. For $m = 0$, $\alpha = 0$. For $n = 0$, $\alpha = 1$. Otherwise, $\alpha$ can be any real.

**b. For fixed values of $m$ and $n$, what value of $α$ yields an unbiased estimator with minimum variance?**

When $m$ or $n$ is zero, $\alpha$ is constrained to a specific value. So, this is the minimum variance unbiased choice of $\alpha$.

Consider the case where $m \gt 0$, $n \gt 0$.

Take the variance.

$$
var[\hat{\theta}] = var[\alpha\overline{X}_m + (1 - \alpha)\overline{Y}_n]
$$

For fixed $m$, $n$, $\alpha$, the quantitites $\alpha\overline{X}_m$ and $(1 - \alpha)\overline{Y}_n$ are independent.

$$
var[\hat{\theta}] = var[\alpha\overline{X}_m] + var[(1 - \alpha)\overline{Y}_n]
$$

$$
var[\hat{\theta}] = var[\alpha{1 \over m}\sum_{i = 1}^m X_i] + var[(1 - \alpha) {1 \over n}\sum_{i = 1}^n Y_i]
$$

$$
var[\hat{\theta}] = (\alpha/m)^2var[\sum_{i = 1}^m X_i] + ((1 - \alpha)/n)^2 var[\sum_{i = 1}^n Y_i]
$$

The $X_i$ and $Y_i$ are random samples, so they are independent.

$$
var[\hat{\theta}] = (\alpha/m)^2\sum_{i = 1}^m var[X_i] + ((1 - \alpha)/n)^2 \sum_{i = 1}^n var[Y_i]
$$

$$
var[\hat{\theta}] = (\alpha/m)^2\sum_{i = 1}^m 4\phi + ((1 - \alpha)/n)^2 \sum_{i = 1}^n \phi
$$

$$
var[\hat{\theta}] = (\alpha/m)^24m\phi + ((1 - \alpha)/n)^2 n \phi
$$

$$
var[\hat{\theta}] = {4m\alpha^2 \over m^2} + {(\alpha^2 -2\alpha + 1)n \over n^2}
$$

$$
var[\hat{\theta}] = {4n\alpha^2 \over mn} + {(\alpha^2 -2\alpha + 1)m \over mn}
$$

$$
var[\hat{\theta}] = {(4n + m)\alpha^2 - 2m\alpha + m \over mn}
$$

We are interested in the minimum variance estimator. So, we will do gradient-based optimization.

$$
{d \over d\alpha} var[\hat{\theta}] = {d \over d\alpha}{(4n + m)\alpha^2 - 2m\alpha + m \over mn}
$$

$$
{d \over d\alpha} var[\hat{\theta}] = {2(4n + m)\alpha - 2m \over mn}
$$

$$
{d^2 \over d^2\alpha} var[\hat{\theta}] = {d \over d\alpha}{2(4n + m)\alpha - 2m \over mn}
$$

$$
{d^2 \over d^2\alpha} var[\hat{\theta}] = {8n + 2m \over mn}
$$

$$
{d^2 \over d^2\alpha} var[\hat{\theta}] = {6n + 2(m + n) \over mn}
$$

So,

$$
{d^2 \over d^2\alpha} var[\hat{\theta}] \gt 0
$$

Suppose,

$$
{d \over d\alpha} var[\hat{\theta}] = 0
$$

Then,

$$
{2(4n + m)\alpha - 2m \over mn} = 0
$$

$$
2(4n + m)\alpha = 2m
$$

So, there is a critical point at.

$$
\alpha = {m \over 4n + m}
$$

The second derivative is positive. So, this is a minimum.

So, the choice of $\alpha$ depends on $m$ and $n$. For $m = 0$, $\alpha = 0$. For $n = 0$, $\alpha = 1$. Otherwise, $\alpha = {m \over 4n + m}$.
