**2. It is not known what proportion p of the purchases of a certain brand of breakfast cereal are made by women and what proportion are made by men. In a random sample of
70 purchases of this cereal, it was found that 58 were made by women and 12 were made by men. Find the M.L.E. of p.**

Let $M$ be the random variable representing the number of purchases made by men in the sample. Then, under the assumption that the sample is iid, we can use the model,

$$
M \sim \text{Binomial}(70, p)
$$

Then,

$$
P(M = m | p) = {70 \choose m}p^{m}(1 - p)^{70 - m}
$$

The likelihood $L$ can be any function that satisfies,

$$
L(p) \propto P(M = 12 | p)
$$

Let,

$$
L(p) = p^{12}(1 - p)^{58}
$$

Then,

$$
\ln(L(p)) = \ln(p^{12}(1 - p)^{58})
$$

$$
\ln(L(p)) = \ln(p^{12}) + \ln((1 - p)^{58})
$$

$$
\ln(L(p)) = 12\ln(p) + 58\ln(1 - p)
$$

Take the derivative.

$$
{d \over dp} \ln(L(p)) = {d \over dp} \left( 12\ln(p) + 58\ln(1 - p) \right)
$$

$$
{d \over dp} \ln(L(p)) = 12{d \over dp} \ln(p) + 58{d \over dp}\ln(1 - p)
$$

$$
{d \over dp} \ln(L(p)) = {12 \over p} - {58 \over 1 - p}
$$

We are interested in $p$ for which ${d \over dp} \ln(L(p)) = 0$.

$$
{12 \over p} - {58 \over 1 - p} = 0
$$

$$
{12 \over p} = {58 \over 1 - p}
$$

$$
12(1 - p) = 58p
$$

$$
12 - 12p = 58p
$$

$$
12 = 70p
$$

$$
p = {12 \over 70} \approx 0.171
$$

Since ${d \over dx} \ln x \gt 0$, this must be a critical point of $L$.

Optimal values must be achieved at critical points or boundary conditions.

$$
L(0) = (0)^{12}(1)^{58} = 0
$$

$$
L(1) = (1)^{12}(0)^{58} = 0
$$

$$
L(0.171) = (0.171)^{12}(1 - (0.171))^{58} \gt 0
$$

So, 0.171 is the maximum likelihood estimate of $p$.

**5. Suppose that X1, . . . , Xn form a random sample from a Poisson distribution for which the mean θ is unknown, (θ > 0).**

**a. Determine the M.L.E. of θ, assuming that at least one of the observed values is different from 0.**

We have that,

$$
X_i \sim \text{Poisson}(\theta)
$$

So,

$$
P(X_i = x_i | \theta) = {\theta^{x_i} \exp(-\theta) \over x_i!}
$$

The $X_i$ are iid, so the joint distribution factorizes as,

$$
P(X_1...X_n | \theta) = \prod_{i = 1}^n {\theta^{x_i} \exp(-\theta) \over x_i!}
$$

The likelihood $L$ is any function that satisfies,

$$
L(\theta) \propto P(X_1...X_n | \theta)
$$

Let,

$$
L(\theta) = \prod_{i = 1}^n \theta^{x_i}\exp(-\theta)
$$

Then,

$$
\ln(L(\theta)) = \ln \left(\prod_{i = 1}^n \theta^{x_i}\exp(-\theta) \right)
$$

$$
\ln(L(\theta)) = \sum_{i = 1}^n \bigg( \ln (\theta^{x_i}) + \ln(\exp(-\theta)) \bigg)
$$

$$
\ln(L(\theta)) = \sum_{i = 1}^n \bigg( x_i \ln (\theta) -\theta \bigg)
$$

Let,

$$
s = \sum_{i = 1}^n x_i
$$

Then,

$$
\ln(L(\theta)) = \ln (\theta)s - n\theta
$$

Then,

$$
{d \over d\theta}\ln(L(\theta)) = {s \over \theta} - n
$$

We are interested in critical points. So, set,

$$
{s \over \theta} - n = 0
$$

$$
{s \over \theta} = n
$$

$$
\theta = {s \over n}
$$

$$
\theta = {1 \over n}\sum{i = 1}^n x_i
$$

$$
\theta = \bar{X}_n
$$

It is an assumption that $\theta$ lies on an open set, so if a maximum of $\ln(L)$ exists it must be this one. Since $\ln$ is monotonic increasing, this is also the maximum likelihood estimator of $\theta$. So, the MLE of $\theta$ is the sample mean.

**b. Show that the M.L.E. of θ does not exist if every observed value is 0.**

If every observed value is zero, then the sample mean is zero. The parameter of the poisson distribution cannot be zero, so the MLE does not exist.

The reason that the parameter of the poisson distribution cannot be zero is that then the density would be zero everywhere, and it would be an invalid distribution. The likelihood is convex and maximized at zero, but it is constrained to positive reals. So, the optimum value is the smallest positive real, which does not exist.

**7. Suppose that X1, . . . , Xn form a random sample from an exponential distribution for which the value of the parameter β is unknown (β > 0). Find the M.L.E. of β.**

We have that,

$$
X_i \sim \text{Exponential}(\beta)
$$

So,

$$
P(X_i = x_i | \beta) = \beta\exp(-\beta x_i)
$$

The $X_i$ are iid, so the joint distribution factorizes as,

$$
P(X_1 = x_1...X_n = x_n | \beta) = \prod_{i = 1}^n \beta\exp(-\beta x_i)
$$

Let,

$$
L(\theta) = \prod_{i = 1}^n \beta\exp(-\beta x_i)
$$

Then,

$$
\ln(L(\theta)) = \ln \left( \prod_{i = 1}^n \beta\exp(-\beta x_i) \right)
$$

$$
\ln(L(\theta)) =  \sum_{i = 1}^n \ln \left(\beta\exp(-\beta x_i) \right)
$$

$$
\ln(L(\theta)) =  \sum_{i = 1}^n \left( \ln (\beta) + \ln(\exp(-\beta x_i)) \right)
$$

$$
\ln(L(\theta)) =  \sum_{i = 1}^n \left( \ln (\beta) - \beta x_i \right)
$$

$$
\ln(L(\theta)) =  \sum_{i = 1}^n \ln (\beta) - \sum_{i = 1}^n \beta x_i
$$

$$
\ln(L(\theta)) =  n\ln (\beta) - \beta \sum_{i = 1}^n x_i
$$

Let,

$$
s = \sum_{i = 1}^n x_i
$$

Then,

$$
\ln(L(\theta)) =  n\ln (\beta) - \beta s
$$

$$
{d \over d\beta}\ln(L(\theta)) =  {d \over d\beta}\left( n\ln (\beta) - \beta s \right)
$$

$$
{d \over d\beta}\ln(L(\theta)) =  {n \over \beta} - s
$$

We are interested in critical points.

$$
{d \over d\beta}\ln(L(\theta)) =  0
$$

$$
{n \over \beta} - s = 0
$$

$$
{n \over \beta} = s
$$

$$
\beta = {n \over s}
$$

$$
\beta = \left({1 \over n} \sum_{i = 1}^n x_i \right)^{-1}
$$

$$
\beta = \bar{X}_n^{-1}
$$

So, the maximum likelihood estimator of $\beta$ is the reciprocal of the sample mean.

**9. Suppose that X1, . . . , Xn form a random sample from a distribution for which the p.d.f. f (x|θ) is as follows:**

$$
f(x) = \begin{cases}
\theta x^{\theta - 1} & 0 \lt x \lt 1 \\
0 & \text{otherwise}
\end{cases}
$$

**Also, suppose that the value of θ is unknown (θ > 0). Find the M.L.E. of θ.**

Assuming all $x_i \in (0, 1)$, we get,

$$
L(\theta) = P(x_i...x_n | \theta) = \prod_{i = 1}^n \theta x_i^{\theta - 1}
$$

Then,

$$
\ln(L(\theta)) = \ln \left(\prod_{i = 1}^n \theta x_i^{\theta - 1} \right)
$$

$$
\ln(L(\theta)) = \sum_{i = 1}^n \left( \ln(\theta) + (\theta - 1)\ln(x_i) \right)
$$

$$
\ln(L(\theta)) = \sum_{i = 1}^n \left( \ln(\theta) + \theta\ln(x_i) - \ln(x_i) \right)
$$

Take the derivative,

$$
{d \over d\theta}\ln(L(\theta)) = {d \over d\theta}\sum_{i = 1}^n \left( \ln(\theta) + \theta\ln(x_i) - \ln(x_i) \right)
$$

$$
{d \over d\theta}\ln(L(\theta)) = \sum_{i = 1}^n \left( {1 \over \theta} + \ln(x_i) \right)
$$

$$
{d \over d\theta}\ln(L(\theta)) = {n \over \theta} + \sum_{i = 1}^n \ln(x_i)
$$

Set,

$$
{d \over d\theta}\ln(L(\theta)) = 0
$$

Then,

$$
{n \over \theta} + \sum_{i = 1}^n \ln(x_i) = 0
$$

$$
n + \theta\sum_{i = 1}^n \ln(x_i) = 0
$$

$$
\theta\sum_{i = 1}^n \ln(x_i) = -n
$$

$$
\theta = -{n \over \sum_{i = 1}^n \ln(x_i)}
$$

So, when $n = 0$, $\theta = 0$. As, this is precluded by the bound on $\theta$, the MLE does not exist in this case. Otherwise,

$$
\theta = -\left( {1 \over n}\sum_{i = 1}^n \ln(x_i) \right)^{-1}
$$

The $X_i$ are bounded exclusive by 0 and 1, so this is always a positive number. Since $\theta$ is constrained to an open set, if a maximum lieklihood estimator exists, it will be given by this formula.

**11. Suppose that X1, . . . , Xn form a random sample from the uniform distribution on the interval [θ1, θ2], where both θ1 and θ2 are unknown (−∞ < θ1 < θ2 < ∞). Find the M.L.E.’s of θ1 and θ2.**

The joint distribution factorizes as,

$$
P(X_1=x_1,...X_n = x_n | \theta_1, \theta_2) = \prod_{i = 1}^n {1 \over \theta_2 - \theta_1} I(\theta_1 \lt x_i \lt \theta_2)
$$

We'll take this as the likelihood. So,

$$
P(X_1=x_1,...X_n = x_n | \theta_1, \theta_2) = I(\theta_1 \lt \min(x_i) \le \max(x_i) \lt \theta_2) \prod_{i = 1}^n {1 \over \theta_2 - \theta_1} 
$$

$$
P(X_1=x_1,...X_n = x_n | \theta_1, \theta_2) = I(\theta_1 \lt \min(x_i) \le \max(x_i) \lt \theta_2) (\theta_2 - \theta_1)^{-n}
$$

So, subject to the constraint that $x_i \in [\theta_1, \theta_2]$, the likelihood is maximized when the difference $\theta_2 - \theta_1$ is minimized. So, the MLE is $\theta_1 = \min(x_1...x_n)$ and $\theta_2 = \max(x_1...x_n)$.

This has a couple issues. First, if $n = 1$, then $\theta_1 = \theta_2$. So, either the MLE is not defined or its a delta function, neither of which is aethetically pleasing. Second, the MLE always assigns zero probability to $x_i$ outside the bounds of what's been seen before, which does not seem like reasonable behavior. For these reasons, this example and similar ones are sometimes cited as reasons to not always use MLE.

**2. Suppose that X1, . . . , Xn form a random sample from a Poisson distribution for which the mean is unknown. Determine the M.L.E. of the standard deviation of the distribution.**

For the poisson distribution, the standard deviation is equal to the square root of the mean. Since both the mean and standard deviation are positive, this function is 1:1 and onto. So, by invariance, the MLE of the standard deviation is the square root of the MLE of the mean. Above, we found that the MLE of the mean for the poisson distribution is the sample mean when this number is greater than zero. So, the MLE of the standard deviation is the sqaure root of the sample mean when this number is greater than 0.

**6. Suppose that X1, . . . , Xn form a random sample from a normal distribution for which both the mean and the variance are unknown. Find the M.L.E. of the 0.95 quantile of the distribution, that is, of the point θ such that $Pr(X < θ)= 0.95$.**

Let $\Phi(x)$ be the cdf function of the standard normal variable, and let $q = 1.6448536269514722$. Then, $\Phi(q) \approx 0.95$.

The mean, $\mu$, and variance, $\sigma^2$, of the normal distribution uniquely determine the 95th quantile, $q^*(\mu, \sigma^2)$.

$$
q^*(\mu, \sigma^2) = \mu + q\sqrt{\sigma^2}
$$

Let $\hat{\mu}$ and $\hat{\sigma^2}$ represent the MLE estimators for the mean and variance respectively, and let $\hat{q}$ represent the MLE estimator of the 95th quantile. Then, by invariance,

$$
\hat{q} = \hat{\mu} + q\sqrt{\hat{\sigma^2}}
$$

The MLE estimators of the mean and variance are the sample mean $\overline{X_n}$ and unadjusted sample variance $\overline{X^2_n}$ respectively. So,

$$
\hat{q} = \overline{X_n} + q\sqrt{\overline{X^2_n}}
$$

**10. Suppose that X1, . . . , Xn form a random sample from a beta distribution for which both parameters α and β are unknown. Show that the M.L.E.’s of α and β satisfy the following equation**

$$
{\Gamma'(\hat{\alpha}) \over \Gamma(\hat{\alpha})}
- {\Gamma'(\hat{\beta}) \over \Gamma(\hat{\beta})} =
{1 \over n}\sum_{i = 1}^n \log{X_i \over 1 - X_i}
$$

Assume $x_i \in (0, 1)$ which is an open set, $n \gt 0$, $\alpha \gt 0$ and $\beta \gt 0$. Then, the likelihood is given by,

$$
L(\alpha, \beta) = P(X_1...X_n | \alpha, \beta) = \prod_{i = 1}^n{\Gamma(\alpha + \beta) \over \Gamma(\alpha) \Gamma(\beta)} x_i^{\alpha - 1}(1 - x_i)^{\beta - 1}
$$

Then, using the assumption that $x_i \not\in \{0, 1\}$ the log likelihood is given by,

$$
\ln(L(\alpha, \beta)) = \ln \left( \prod_{i = 1}^n{\Gamma(\alpha + \beta) \over \Gamma(\alpha) \Gamma(\beta)} x_i^{\alpha - 1}(1 - x_i)^{\beta - 1} \right)
$$

$$
\ln(L(\alpha, \beta)) = \sum_{i = 1}^n \ln\left({\Gamma(\alpha + \beta) \over \Gamma(\alpha) \Gamma(\beta)} x_i^{\alpha - 1}(1 - x_i)^{\beta - 1} \right)
$$

$$
\ln(L(\alpha, \beta)) = \sum_{i = 1}^n \left(
\ln {\Gamma(\alpha + \beta) \over \Gamma(\alpha) \Gamma(\beta)}
+ \ln(x_i^{\alpha - 1})
+ \ln((1 - x_i)^{\beta - 1})
\right)
$$

$$
\ln(L(\alpha, \beta)) = \sum_{i = 1}^n(
    \ln(\Gamma(\alpha + \beta))
    - \ln(\Gamma(\alpha))
    - \ln(\Gamma(\beta))
    + (\alpha - 1)\ln(x_i)
    + (\beta - 1)\ln(1 - x_i)
)
$$

$$
\ln(L(\alpha, \beta)) =
n\ln(\Gamma(\alpha + \beta))
- n\ln(\Gamma(\alpha))
- n\ln(\Gamma(\beta))
+ (\alpha - 1)\sum_{i = 1}^n \ln(x_i)
+ (\beta - 1)\sum_{i = 1}^n\ln(1 - x_i)
$$

Take the partial derivatives,

$$
{d \over d\alpha}
\ln(L(\alpha, \beta)) =
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
- n{\Gamma'(\alpha) \over \Gamma(\alpha)}
+ \sum_{i = 1}^n \ln(x_i)
$$

$$
{d \over d\beta}
\ln(L(\alpha, \beta)) =
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
- n{\Gamma'(\beta) \over \Gamma(\beta)}
+ \sum_{i = 1}^n \ln(1 - x_i)
$$



Set these to zero,

$$
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
- n{\Gamma'(\alpha) \over \Gamma(\alpha)}
+ \sum_{i = 1}^n \ln(x_i)
= 0
$$

$$
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
- n{\Gamma'(\beta) \over \Gamma(\beta)}
+ \sum_{i = 1}^n \ln(1 - x_i)
= 0
$$

So,

$$
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
+ \sum_{i = 1}^n \ln(x_i)
= n{\Gamma'(\alpha) \over \Gamma(\alpha)}
$$

$$
n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
+ \sum_{i = 1}^n \ln(1 - x_i)
= n{\Gamma'(\beta) \over \Gamma(\beta)}
$$

Subtract,

$$
\left(
    n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
    + \sum_{i = 1}^n \ln(x_i)
\right)-
\left(
    n{\Gamma'(\alpha + \beta) \over \Gamma(\alpha + \beta)}
    + \sum_{i = 1}^n \ln(1 - x_i)
\right)
=
\left(
    n{\Gamma'(\alpha) \over \Gamma(\alpha)}
\right)-
\left(
    n{\Gamma'(\beta) \over \Gamma(\beta)}
\right)
$$

Then,

$$
\sum_{i = 1}^n \ln(x_i)
- \sum_{i = 1}^n \ln(1 - x_i)
=
n{\Gamma'(\alpha) \over \Gamma(\alpha)}
- n{\Gamma'(\beta) \over \Gamma(\beta)}
$$

$$
{1 \over n} \sum_{i = 1}^n \left( \ln(x_i) - \ln(1 - x_i) \right)
=
{\Gamma'(\alpha) \over \Gamma(\alpha)}
- {\Gamma'(\beta) \over \Gamma(\beta)}
$$

$$
{1 \over n} \sum_{i = 1}^n \ln \left({x_i \over 1 - x_i}\right)
=
{\Gamma'(\alpha) \over \Gamma(\alpha)}
- {\Gamma'(\beta) \over \Gamma(\beta)}
$$

Since $\alpha$ and $\beta$ are constrained to open sets, if there exists an MLE for each, then it is achieved when the gradient of the joint likelihood function is zero. This equation was derived under the assumption that the gradient is zero. So, if the MLE estimators exist for these parameters, then,

$$
{1 \over n} \sum_{i = 1}^n \ln \left({x_i \over 1 - x_i}\right)
=
{\Gamma'(\hat{\alpha}) \over \Gamma(\hat{\alpha})}
- {\Gamma'(\hat{\beta}) \over \Gamma(\hat{\beta})}
$$
